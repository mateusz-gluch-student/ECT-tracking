```python
from ect import logpolar, Config

import cv2
from matplotlib import pyplot as plt
import numpy as np

from loguru import logger
# import sys
# logger.add(sys.stderr, level="DEBUG")
```

### Load sample image

```python
SAMPLE_IMAGE = "./sample_images/smile.png"

src = cv2.imread(SAMPLE_IMAGE)
src = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)

plt.imshow(src)
```

### Sampling modes testing

```python
# bilinear
radius = 100
center = (300, 100)
dsize = (314, 100)

dst = logpolar(src, center, dsize, radius)

plt.imshow(dst)
```

```python
# bilinear
radius = 100
center = (100, 100)
dsize = (314, 100)

cfg = Config(logpolar_mode="omit")

dst = logpolar(src, center, dsize, radius, cfg)

plt.imshow(dst)
```

```python
# bilinear
radius = 100
center = (100, 100)
dsize = (314, 100)

cfg = Config(logpolar_mode="opencv")
dst = logpolar(src, center, dsize, radius, cfg)

plt.imshow(dst)
```
