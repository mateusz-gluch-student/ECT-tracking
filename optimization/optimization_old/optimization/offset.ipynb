```python
import ect
import matplotlib.pyplot as plt
import numpy as np
import cv2

from ect.optimizers import ECTOffsetOptimizer

global_radius = 200
global_offset = 20
```

```python
source_path = "../sample_imgs/black_e.png"

source_img = cv2.imread(source_path)
source_img = cv2.cvtColor(source_img, cv2.COLOR_BGR2GRAY)
source_img = cv2.resize(source_img, None, None, fx=0.5, fy=0.5)

plt.imshow(source_img)
```

```python
def rms(x: np.ndarray) -> float:
    rms_ = np.sqrt((x*np.conj(x)).sum().sum())
    return rms_

def rmsnorm(x: np.ndarray) -> np.ndarray:
    rms = np.sqrt((x*np.conj(x)).sum().sum())
    return x/rms

def snr(image: cv2.Mat, template: cv2.Mat):
    snr_lin = (1/rms((rmsnorm(image) - rmsnorm(template))))
    return 20*np.log(snr_lin)
```

```python
blacke = ECTOffsetOptimizer(
    image_path = "../sample_imgs/black_e.png",
    radius = global_radius,
    img_offset = global_offset,
    loss_fcn = lambda x, y: -snr(x, y)
)

print(f"{blacke.loss([45])=}")
print(f"{blacke.loss_sidelobe([45])=}")

blacke.show_result(rmsnorm, 45)
```

```python
result = blacke.optim(method = "Nelder-Mead")

solution = result['x'][0]
```

```python
blacke.show_result(solution, rmsnorm)

print(f"{blacke.loss([solution])=}")
print(f"{blacke.loss_sidelobe([solution])=}")
```
